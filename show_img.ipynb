```python
# Visual inspection of query and nearest neighbors from get_paths_eval.py
fig, axes = plt.subplots(1, 6, figsize=(15,5))
query_img = Image.open("/shared/data/Hotels-50K/images/val_query/64/8172/travel_website/7473757.jpg")

axes[0].imshow(query_img)
axes[0].set_title("Query")
axes[0].axis('off')

test_paths = ["/shared/data/Hotels-50K/images/val/61/18536/travel_website/6565762.jpg",
              "/shared/data/Hotels-50K/images/val/64/48306/travel_website/7194571.jpg",
             "/shared/data/Hotels-50K/images/val/92/38744/travel_website/6613073.jpg",
             "/shared/data/Hotels-50K/images/val/92/83858/travel_website/7960893.jpg",
             "/shared/data/Hotels-50K/images/val/92/35676/travel_website/7064041.jpg"]

for i, path in enumerate(test_paths):
    img = Image.open(path)
    axes[i+1].imshow(img)
    axes[i+1].set_title(f"Neighbor {i+1}")
    axes[i+1].axis('off')
    
plt.tight_layout()
plt.show()
```
