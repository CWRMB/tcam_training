```python
# Visual inspection of query and nearest neighbors from get_paths_eval.py
fig, axes = plt.subplots(1, 6, figsize=(15,5))
query_img = Image.open("/shared/data/Hotels-50K/images/val_query/91/124897/travel_website/8992072.jpg")

axes[0].imshow(query_img)
axes[0].set_title("Query")

test_paths = ["/shared/data/Hotels-50K/images/val/91/124897/travel_website/8992080.jpg",
              "/shared/data/Hotels-50K/images/val/91/27063/travel_website/6574661.jpg",
             "/shared/data/Hotels-50K/images/val/91/8706/travel_website/6104827.jpg",
             "/shared/data/Hotels-50K/images/val/91/55449/travel_website/6096471.jpg",
             "/shared/data/Hotels-50K/images/val/92/4150/travel_website/6325195.jpg"]

for i, path in enumerate(test_paths):
    img = Image.open(path)
    axes[i+1].imshow(img)
    axes[i+1].set_title(f"Neighbor {i+1}")
    
plt.tight_layout()
plt.show()
```
